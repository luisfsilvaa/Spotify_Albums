# Imports 

```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
```

# Load Data

```python
albums = pd.read_csv('eda_albums.csv', index_col=0)
albums['descs'] = albums['descs'].fillna('')
```

```python
albums.head()
```

# NLP Preprocessing

```python
df = albums.drop(columns=['num_rat', 'num_revs', 'acousticness', 'energy', 'instrumentalness', 'danceability', 'liveness', 'loudness', 'speechiness','tempo', 'valence', 'duration_ms', 'time_signature'])
```

```python
df.head()
```

```python
df['name'] = df['ars_name'] + ' - ' + df['album']
df['text'] = df['gens'] + ' ' + df['descs']
df['weight_text'] = 2*(df['gens'] + ' ') + df['descs']
```

```python
df['text'] = df['text'].str.replace(',','').str.lower()
df['weight_text'] = df['weight_text'].str.replace(',','').str.lower()
```
